```python
import json
import pandas as pd
```

```python
with open('movie_to_relevant_tags.json', 'r') as fp:
        movie_to_tags = json.load(fp)
```

```python
movie_to_tags["1"]
```

```python
movies_df = pd.read_csv("./original-dataset/movies.csv")
movies_df
```

```python
movies = {}

for i, row in movies_df.iterrows():
    movie_id = str(row["movieId"])

    tags = None

    if movie_id in movie_to_tags:
        tags = [x.capitalize() for x in movie_to_tags[movie_id]]
    else:
        tags = ["No tags"]

    movies[movie_id] = {
        "title": row["title"],
        "genres": row["genres"].split("|"),
        "tags": tags
    }

    print(i, end="\r")
```

```python
movies
```

```python
with open("movies.json", "w") as outfile:
    json.dump(movies, outfile)
```
